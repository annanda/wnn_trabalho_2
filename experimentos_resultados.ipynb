```python
import csv
import matplotlib.pyplot as plt

example = []
log_loss = []
with open('dog_examples.csv', 'rb') as csv_file:
    spamreader = csv.reader(csv_file, delimiter=',')
    for row in spamreader:
        example.append(row[0])
        log_loss.append(row[1])
```

```python
plt.figure(figsize=(15,5))
plt.title("Log loss em cachorros")
plt.plot(example[], log_loss, '-b', label="log loss")
plt.plot(example[28], log_loss[28], '*g', label="menor log loss")
plt.plot(example[44], log_loss[44], '*r', label="limite exemplos")
plt.legend(loc=0) ; plt.show()
```
